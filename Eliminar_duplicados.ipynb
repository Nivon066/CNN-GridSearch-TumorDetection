```python
import hashlib
import os
base_dir = "/content/drive/MyDrive/Inteligencia Artificial/Datasets/Brain_Tumor_DS"
labels = ['glioma', 'meningioma','notumor','pituitary']
```

```python
def compute_hash(file):
    hasher = hashlib.md5()
    with open(file, 'rb') as f:
        buf = f.read()
        hasher.update(buf)
    return hasher.hexdigest()
```

```python
def list_files(base_dir, labels, hash_dict):
    for data_type in ['Training', 'Testing']:
        for label in labels:
            folder_path = os.path.join(base_dir, data_type, label)
            for root, dirs, files in os.walk(folder_path):
                for file in files:
                    if file.endswith(".jpg"):
                        file_path = os.path.join(root, file)
                        file_hash = compute_hash(file_path)
                        if file_hash in hash_dict:
                            hash_dict[file_hash].append(file_path)
                        else:
                            hash_dict[file_hash] = [file_path]

```

```python
def remove_duplicates(hash_dict):
    duplicate_count = 0
    for hash_value, file_paths in hash_dict.items():
        if len(file_paths) > 1:
            for file_path in file_paths[1:]:
                print(f"Removing duplicate (hash : {hash_value}) : {file_path}")
                os.remove(file_path)
                duplicate_count += 1
    print(f"Number of duplicates : {duplicate_count}")
```

```python
if __name__ == '__main__':
    hash_dict = {}
    list_files(base_dir, labels, hash_dict)
    remove_duplicates(hash_dict)
```
